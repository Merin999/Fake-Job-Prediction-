```python
import numpy as np
import pandas as pd
import joblib
```

```python
df=pd.read_csv('cleaned_data.csv')
print(df.head())
```

```python
def find_category_mappings(df, variable):
    return {k: i for i, k in enumerate(df[variable].unique())}
def integer_encode(finaldf,variable, ordinal_mapping):
    df[variable+"_new"] = df[variable].map(ordinal_mapping)
for variable in ['location','employment_type','required_experience','required_education','industry','function']:
    mappings = find_category_mappings(df,variable)
    file=open(variable+".txt",'w')
    file.write(str(mappings))
    file.close()
    integer_encode(df, variable, mappings)
df.head()
```

```python
X=df.drop(['job_id','cleaned_text','text','location','employment_type','required_experience','required_education','industry','function','fraudulent','title','department','salary_range','company_profile','description','requirements','benefits'], axis=1)
X.head()
```

```python
X['text_length']
```

```python
y=df['fraudulent']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=50, stratify=y)
```

```python
X_train1=X_train[y_train==0].iloc[0::3].append(X_train[y_train==1])
#print(X_train1)
X_train2=X_train[y_train==0].iloc[1::3].append(X_train[y_train==1])
#print(X_train2)
X_train3=X_train[y_train==0].iloc[2::3].append(X_train[y_train==1])
#print(X_train3)
```

```python
y_train1=y_train[y_train==0].iloc[0::3].append(y_train[y_train==1])
#print(y_train1)
y_train2=y_train[y_train==0].iloc[1::3].append(y_train[y_train==1])
#print(y_train2)
y_train3=y_train[y_train==0].iloc[2::3].append(y_train[y_train==1])
#print(y_train3)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(sampling_strategy=0.6, random_state= 42)
X_train1_resampled, y_train1_resampled = ros.fit_resample(X_train1, y_train1)
X_train2_resampled, y_train2_resampled = ros.fit_resample(X_train2, y_train2)
X_train3_resampled, y_train3_resampled = ros.fit_resample(X_train3, y_train3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC

#vectorizer = CountVectorizer(ngram_range=(1,2)) # looking for both unigrams and bigrams
clf_svm = LinearSVC(C=0.01, class_weight='balanced', random_state=42)
clf_svm.fit(X_train1_resampled, y_train1_resampled)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(max_depth=2, random_state=0)
clf_rf.fit(X_train1_resampled, y_train1_resampled)
```

```python
#y_pred_svm=clf_svm.predict(X_test.iloc[0:1,:])
#np.max(y_pred_svm)
y_pred_svm=clf_svm.predict(X_test)
y_pred_rf=clf_rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("\nsvm\n")
print("accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
print("\nrandom forest\n")
print("accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```
